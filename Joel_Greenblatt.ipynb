# **Importando bibliotecas**

```python
import numpy as np
import pandas as pd
import string
import warnings
import requests
from bs4 import BeautifulSoup
warnings.filterwarnings('ignore')
```

# **Importando tabela do Fundamentus**

```python
url = 'https://www.fundamentus.com.br/resultado.php'
header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"}
r = requests.get(url, headers=header)
df = pd.read_html(r.text, decimal=',', thousands='.')[0]
display(df)
```

# **Ajustando decimais de porcentagem**

```python
for coluna in ['Div.Yield', 'Mrg Ebit', 'Mrg. Líq.', 'ROIC', 'ROE', 'Cresc. Rec.5a']:
  df[coluna] = df[coluna].str.replace('.', '')
  df[coluna] = df[coluna].str.replace(',', '.')
  df[coluna] = df[coluna].str.rstrip('%').astype('float') / 100
```

# **Filtrando de acordo com os parametros do Joel Greenblatt adaptadas ao mercado brasileiro.**

```python
df = df[df['Liq.2meses'] > 200000]
df = df[df['EV/EBIT'] > 0]
df = df[df['ROIC'] > 0]
display(df)
```

# **Criando o ranking conforme o posicionamento de cada ativo pressuposto pela Magic Formula**

```python
#Criando o dataframe
ranking = pd.DataFrame()

#Criando um range que vai até o número de linhas existentes na DF
ranking['POS'] = range(1, 235)

#Colocando a coluna EV/EBIT de maneira crescente
ranking['EV/EBIT'] = df.sort_values(by=['EV/EBIT'])['Papel'][:235].values

#Colocando a coluna ROIC de maneira decrescente
ranking['ROIC'] = df.sort_values(by=['ROIC'], ascending=False)['Papel'][:235].values

#Colocando a coluna Div.Yield de maneira descrescente 
ranking['Div.Yield'] = df.sort_values(by=['Div.Yield'], ascending=False)['Papel'][:235].values

#Mostrando como ficou o ranking
display(ranking)
```

```python
a = ranking.pivot_table(columns='EV/EBIT', values='POS')
b = ranking.pivot_table(columns='ROIC', values='POS')
c = ranking.pivot_table(columns='Div.Yield', values='POS')
t = pd.concat([a, b, c])
rank = t.dropna(axis=1).sum()
display(rank.sort_values()[:30])
```

# **Vamos definir quem são nossos ativos**

```python
ativos = rank.sort_values()[:30]
```

# **Define a URL padrão de onde queremos coletar os setores**

```python
url = 'https://statusinvest.com.br/acoes/'
```

# **Criando um USER_AGENTS aleatorio para não dar erro 403 no Requests**

```python
from random import choice

USER_AGENTS = [
    'Mozilla/5.0 (CrKey armv7l 1.5.16041) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/31.0.1650.0 Safari/537.36',
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/'
    '42.0.2311.135 Safari/537.36 Edge/12.246',
    'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/'
    '9.0.2 Safari/601.3.9',
    'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:15.0) Gecko/20100101 Firefox/15.0.1',
    'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.111 Safari/537.36'
]

def get_random_user_agent():
    """ Randoms user agent to prevent "python" user agent
    :return: Random user agent from USER_AGENTS
    :rtype: str
    """
    return choice(USER_AGENTS)
```

# **Vamos armazenar os setores coletados em uma lista**

```python
setores = []
```

```python
# Como a URL é padrao, vamos fazer varias requests, uma pra cada ativo 
for ativo in ativos.index:

  # fornece a malandragem no headers da requisição
  response = requests.get(url+ativo, headers={'User-Agent': get_random_user_agent()})

  # a resposta do site é um html, o BeautifulSoup consegue fazer o parsing desse html
  parser = BeautifulSoup(response.content, "html.parser")

  # Como eu te mostrei, aquele bloco esta numa div contendo a classe "info pr-md-2"
  data = parser.find(class_='info pr-md-2')

  # data é o html que contem a div que contem o texto do setor
  # Como eu te mostrei também, a classe dessa div se chama "value"
  # pode ser que ela não exista, depois vc confere pra coletar manualmente ou
  # inspecionar o site em busca da classe que contem o valo especifico pra esse ativo
  if data is not None:
    value = data.find(class_='value')
    # armazena na lista o par que confere o ativo -> setor
    setores.append((ativo, value.text))
```

```python
# Vamos visualizar os ativos e setores num dataframe
ativos_df = pd.DataFrame(setores, columns=['ATIVO', 'SETOR'])
```
